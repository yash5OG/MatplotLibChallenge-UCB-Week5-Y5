## Observations and Insights 

* The data is considerably clean with only 1 duplicate value

* The sex distribution of the population tested is almost even split between male and female (50.4 % vs 49.6%) 

* Based on the box plots observed below, Capomulin seems to be comparable to Ramicane, and both seem to be  more effective than Infubinol and Ceftamine.

* The shape of the scatter and the line suggests a strong correlation between weight of mice and tumor volume, i.e. mice with higher weight also have higher tumor volume. With a correlation coefficient of 0.84, it is a positive relationship

* Slope of 0.95 also indicates high relationship between the weight and tumor volume, suggesting that for 1 unit increase in weight, tumor volume increases with a factor of 0.95



```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

```

```python
# Combine the data into a single dataset
mouse_study_data = pd.merge(mouse_metadata, study_results,on = "Mouse ID", how="outer")

# Display the data table for preview
mouse_study_data.head()
```

```python
# Checking the number of mice.
len(mouse_study_data["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#showing all duplicate values below # Optional: Get all the data for the duplicate mouse ID. 

duplicate_mice = mouse_study_data[mouse_study_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicate_mice
```

```python
len(duplicate_mice["Mouse ID"])

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleandf = mouse_study_data.loc[mouse_study_data["Mouse ID"]!="g989"]
cleandf
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleandf["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

mean = cleandf.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']

median = cleandf.groupby('Drug Regimen').median()['Tumor Volume (mm3)']

variance = cleandf.groupby('Drug Regimen').var()['Tumor Volume (mm3)']

std = cleandf.groupby('Drug Regimen').std()['Tumor Volume (mm3)']

sem = cleandf.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

summary_table = pd.DataFrame({"Mean Tumor Volume":mean,
                             "Median Tumor Volume":median,
                             "Tumor Volume Variance":variance,
                             "Tumor Volume Std. Dev": std,
                             "Tumor Volume Std. Err.": sem})
summary_table



```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_table2 = cleandf.groupby("Drug Regimen").agg({'Tumor Volume (mm3)':["mean","median","var","std","sem"]})
summary_table2
```

## Quartiles, Outliers and Boxplots

```python
# Find max timepoint for each mouse

max_time = pd.DataFrame(cleandf.groupby('Mouse ID')['Timepoint'].max().sort_values()).reset_index().rename(columns={'Timepoint': 'MaxTP'})
max_time
```

```python
# Merge max timepoint onto data_df
boxdata = pd.merge(cleandf, max_time, on='Mouse ID')
boxdata
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Locate each drug from our list 'drugs' in the orginal dataframe

# Put treatments into a list for for loop (and later for plot labels)
drugs = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
drugv = []

for drug in drugs:
    
    df1 = boxdata.loc[boxdata['Drug Regimen'] == drug]

    # Create subset dataframes that narrow down to rows only containing the final timepoint for each regimen
    df2 = df1.loc[df1['Timepoint'] == df1['MaxTP']]
    
    # Create a series that consists of all final volume values for each regimen
    values = df2['Tumor Volume (mm3)']
    drugv.append(values)
    
    # Calculate and print the interquartile range (IQR) for each regimen
    quartiles = values.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'IQR for {drug}: {iqr}')
    
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'Lower Bound for {drug}: {lower_bound}')
    print(f'Upper Bound for {drug}: {upper_bound}')
    
    # Determine outliers using upper and lower bounds
    outliers = (values.loc[(df2['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (df2['Tumor Volume (mm3)'] <= lower_bound)]).count()
    print(f'Number of {drug} outliers: {outliers}\n')

```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
boxchars = dict(marker='*', markerfacecolor='purple', markersize=13.69, markeredgecolor='pink')

# Plot boxplot
plt.boxplot(drugv, flierprops=boxchars)

# Add title and labels to boxplot
plt.title('Final Tumor Volume for each Drug')
plt.ylabel('Final Tumor Volume (mm3)',font='Papyrus')
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(avgs_cap['Weight (g)'],avgs_cap['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")
```

```python
x_values = avgs_cap['Weight (g)']
y_values = avgs_cap['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{corr}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)
```

```python
# Add the linear regression equation and line to plot
fig3, ax3 = plt.subplots(figsize=(13, 10))
plt.scatter(x_values,y_values,s=69, color='violet')
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot of Mouse Weight Versus Average Tumor Volume',font='Papyrus',fontsize =29)
plt.xlabel('Weight(g)',fontsize =13)
plt.ylabel('Average Tumore Volume (mm3)',fontsize =14)
ax3.annotate(line_eq, xy=(17, 39), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',fontsize=29,color="red")

print(f"The r-squared is: {rvalue**2}")

#plt.savefig("../Images/linear_regression.png", bbox_inches = "tight")
plt.show()
```
